```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
boston = pd.read_csv("boston_house_prices.csv")
boston.keys()
```

```python
boston.head()
```

```python
boston.isnull().sum()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})

#bins take min and make and break it in to bins value 1-200 tak 5 5 ki bin bana ligi
#displot ma line nhi hoti yah histrogram show kar nay kliya hai

sns.displot(boston['MEDV'], bins=20)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})

#bins take min and make and break it in to bins value 1-200 tak 5 5 ki bin bana ligi
sns.distplot(boston['MEDV'], bins=30)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})

#bins take min and make and break it in to bins value 1-200 tak 5 5 ki bin bana ligi
sns.histplot(boston['MEDV'], bins=20)
plt.show()
```

```python
correlation_matrix = boston.corr().round(2)
#heatmap is use to identy correlation
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
plt.figure(figsize=(20,5))

features = ['LSTAT', 'RM']
target = boston['MEDV']

for i,col in enumerate(features):
  plt.subplot(1, len(features), i+1)
  x = boston[col]
  y = target
  plt.scatter(x,y, marker='o')
  plt.title(col)
  plt.xlabel(col)
  plt.ylabel('MEDV')
```
